# Cyclotomic sequence

```gap
LoadPackage("num");
```

We start with $S=\langle 5,7,9\rangle$

```gap
s:=NumericalSemigroup(5,7,9);
```

Its associated polynomial is

```gap
x:=Indeterminate(Rationals,"x");;
p:=NumericalSemigroupPolynomial(s,x);;
```

```gap
Print(p);
```

Which is not Kronecker

```gap
IsKroneckerPolynomial(p);
```

```gap
IsCyclotomicNumericalSemigroup(s);
```

Let us compute its cyclotomic exponent sequence

```gap
c:=CyclotomicExponentSequence(NumericalSemigroup(3,5,7),60);
```

Its Betti elements are isolated, and thus a forest

```gap
JupyterSplashDot(DotBinaryRelation(HasseDiagramOfNumericalSemigroup(s,BettiElements(s))));
```

But $\Omega_*\setminus \{3,5,7\}$ is not a forest

```gap
cm:=Filtered(Difference([2..Length(c)],[3,5,7]),i->c[i]<>0);
```

```gap
JupyterSplashDot(DotBinaryRelation(HasseDiagramOfNumericalSemigroup(s,cm)));
```

Now let us see what happens with $\langle 8, 12, 18, 25\rangle$

```gap
s:=NumericalSemigroup(8,12,18,25);
```

```gap
p:=NumericalSemigroupPolynomial(s,x);;
```

```gap
Print(p);
```

```gap
c:=CyclotomicExponentSequence(s,100);
```

```gap
Length(CyclotomicExponentSequence(s,100));
```

```gap
IsKroneckerPolynomial(p);
```

```gap
JupyterSplashDot(DotBinaryRelation(HasseDiagramOfNumericalSemigroup(s,BettiElements(s))));
```

```gap
cm:=Filtered(Difference([2..Length(c)],MinimalGenerators(s)),i->c[i]<>0);
```

```gap
JupyterSplashDot(DotBinaryRelation(HasseDiagramOfNumericalSemigroup(s,cm)));
```

```gap
JupyterSplashDot(DotTreeOfGluingsOfNumericalSemigroup(s));
```
